```python
##tensorflow >2.0
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python

### Vocabulary size
voc_size=10000
```

# One Hot Representation

```python
onehot_repr=[one_hot(words,voc_size)for words in sent] 
print(onehot_repr)
```

# Word Embedding Represntation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
# padding is dine to make every sentense to same lenght
sent_length=8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length) #pre means 0 at font post means 0 at the end
print(embedded_docs)
```

```python
#dimensions are basically no. of features we wanna make
dim =10
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs))
```
